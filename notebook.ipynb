### Importing important libraries :-
- Numpy and Pandas for EDA.
- Matplotlib and Seaborn for Visualization.
- Sklearn for Linear Regression and model evaluation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

### Importing the Dataset

```python
url = 'https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
df = pd.read_csv(url)
df.head()
```

### Visualizing the Dataset using scatterplot

```python
fig = plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Hours', y='Scores')
plt.title('ScatterPlot for Marks scored vs No. of hours')
plt.xlabel('No. of Hours studied')
plt.ylabel('Marks Scored')
```

<div class="alert alert-block alert-info">
Graph shows that there's a somewhat linear relationship between the columns!!
</div>

### Dividing the Dataset into Training-Testing splits.

```python
X = df[['Hours']].values
y = df['Scores'].values

# using 30% of the dataset for testing and rest for training.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

### Training the model 

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

# score values based on the intercept and coeffecient of linear regression model.
yhat_train = lr.coef_ * X_train + lr.intercept_

# visualizing the training data.
fig = plt.figure(figsize=(10,8))
plt.scatter(X_train, y_train)
plt.plot(X_train, yhat_train, '-r')
plt.title('ScatterPlot for training data')
plt.xlabel('No. of hours studied')
plt.ylabel('Marks Scored')
plt.legend(['Actual Score values', 'Predicted line'])
```

<div class="alert alert-block alert-info">
Line fits pretty well on the training data!!
</div>

*Let's check if this is a case of over-fitting by testing on the test data*

### Testing the model

```python
# predicting test values
yhat_test = lr.predict(X_test)

# visualizing the test data
fig = plt.figure(figsize=(8,6))
plt.scatter(X_test, y_test)
plt.plot(X_test, yhat_test, '-r')
plt.title('ScatterPlot for test data')
plt.xlabel('No. of hours studied')
plt.ylabel('Marks Scored')
plt.legend(['Actual Score values', 'Predicted line'])
```

<div class="alert alert-block alert-info">
Model performs pretty well on the test data too!!
</div>

### Evaluating the model

```python
# using r2Score, Mean squared error and Mean absolute error for checking the accuracy of the model.
r_score = r2_score(y_test, yhat_test)
MSE = mean_squared_error(y_test, yhat_test)
MAE = mean_absolute_error(y_test, yhat_test)

# dataframe containing all the values.
eval_df = pd.DataFrame({"Evalution Method": ['R2 score', 'Mean Sqaured Error', 'Mean Absolute Error'], "Values": [r_score, MSE, MAE]})
eval_df
```

<div class="alert alert-block alert-info">
Model has r_square value of $ 0.94 $, which means that model is performing good on the foreign data. </div>

```python
hours_studied = 9.25
pred_score = lr.predict([[hours_studied]])
print(f"Predicted Score if student studies {hours_studied} a day : {pred_score[0]}")
```
